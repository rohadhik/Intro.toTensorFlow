```python
# import the tensorflow library
import tensorflow as tf
print(tf.__version__)
```

```python
# import libraries
import numpy as np
import keras
```

```python
# define the input and output variables
x = np.array([-2.0, -1.0, 2.0, 4.0, 5.0, 7.0])
y = np.array([-13.0, -10.0, -1.0, 5.0, 8.0, 14.0])
```

```python
# initialize the keras sequential model with one dense layer and one input
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# train the model with optimizer and loss 
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# fit the model and train it with 500 epochs
model.fit(x, y, epochs=500)
```

```python
# model summary
model.summary()
```

```python
# predict the new input 
model.predict([9.0])
```

```python
# predict another input
model.predict([40.0])
```

```python
# predict with multiple inputs
model.predict([10.0, 11.0, 15.0])
```

#### This is also called as the Hello world program of the tensorflow keras sequential model.
#### In this we used optimizer 'sgd' to optimize the result w.r.t loss function which is mean_squared_error by reducing the error.
